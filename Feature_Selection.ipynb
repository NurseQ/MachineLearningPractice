```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
```

```python
data = pd.read_csv('Covid Predesposition/covid.csv')
```

```python
data.head()
```

```python
data['cons']=0
```

```python
data.head()
```

```python
data['ones']= 1
```

```python
data.head()
```

```python
data=data.drop(['id','entry_date','date_symptoms','date_died'],axis=1)
data.head()
```

```python
# In all feature selection procedures, it is good practice to select the features by examining only the training set
# To avoid overfitting.

X_train, X_test, y_train, y_test = train_test_split(
    data.drop(labels=['icu'], axis=1),
    data['icu'],test_size=0.3, random_state=3)

X_train.shape, X_test.shape
```

### Using VarianceThreshold

```python
sel = VarianceThreshold(threshold=0)

sel.fit(X_train) # fit finds the features with zero variance
```

```python
# get_support is a boolean vector that indicates which features are retained

sum(sel.get_support()) 
```

```python
constant = X_train.columns[~sel.get_support()]
len(constant)
```

```python
constant
```

```python
for col in constant:
    print(col,X_train[col].unique())
```
