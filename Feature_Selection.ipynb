```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
```

```python
data = pd.read_csv('Covid Predesposition/covid.csv')
```

```python
data.head()
```

```python
data['cons']=0
```

```python
data.head()
```

```python
data['ones']= 1
```

```python
data.head()
```

```python
data=data.drop(['id','entry_date','date_symptoms','date_died'],axis=1)
data.head()
```

```python
# In all feature selection procedures, it is good practice to select the features by examining only the training set
# To avoid overfitting.

X_train, X_test, y_train, y_test = train_test_split(
    data.drop(labels=['icu'], axis=1),
    data['icu'],test_size=0.3, random_state=3)

X_train.shape, X_test.shape
```

### Using VarianceThreshold

```python
sel = VarianceThreshold(threshold=0)

sel.fit(X_train) # fit finds the features with zero variance
```

```python
# get_support is a boolean vector that indicates which features are retained

sum(sel.get_support()) 
```

```python
constant = X_train.columns[~sel.get_support()]
len(constant)
```

```python
constant
```

```python
for col in constant:
    print(col,X_train[col].unique())
```

### Duplicated Features

```python
df = pd.read_csv('Covid Predesposition/covid.csv', nrows=15000)
```

```python
df['cons']=0
```

```python
df['ones']=1
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(labels=['icu'], axis=1),
    df['icu'],test_size=0.3, random_state=3)

X_train.shape, X_test.shape
```

```python
# Transpose df, to be able to use duplicated()
data_t = X_train.T
data_t.head(30)
```

```python
data_t.duplicated().sum()
# this code took too long for original DF, SHOULD BE ONLY USED ON SMALL DF
```

```python
data_t[data_t.duplicated()]
```

```python
# capture duplicated features
dup_features = data_t[data_t.duplicated()].index.values
dup_features
```

```python
# alternatively, remove duplcated rows then transpose
data_uni = data_t.drop_duplicates(keep='first').T
data_uni.shape
```

### Removing duplicates from bigger df

```python
df_big = pd.read_csv('Covid Predesposition/covid.csv')
```

```python
df_big['cons']=0
```

```python
df_big['ones']=1
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df_big.drop(labels=['icu'], axis=1),
    df_big['icu'],test_size=0.3, random_state=3)

X_train.shape, X_test.shape
```

```python
dup_feat = []

for i in range(0, len(X_train.columns)):
    if i % 10 == 0:
        print(i)
    
    col_1 = X_train.columns[i]
    
    for col_2 in X_train.columns[i + 1:]:
        if X_train[col_1].equals(X_train[col_2]):
            dup_feat.append(col_2)
```

```python
print(len(set(dup_feat)))
```
